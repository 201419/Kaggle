```python
# LOAD LIBRARIES
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd, numpy as np, gc
import warnings
warnings.filterwarnings("ignore")
from sklearn.metrics import roc_auc_score
import lightgbm as lgb
import statsmodels.api as sm
```

```python
d = {}
for i in range(200): d['var_'+str(i)] = 'float32'
d['target'] = 'uint8'
d['ID_code'] = 'object'

train = pd.read_csv('./input/train.csv', dtype=d)
test = pd.read_csv('./input/test.csv', dtype=d)

print('Loaded',len(train),'rows of train')
print('Loaded',len(test),'rows of test')
```

```python
train2 = train.sample(frac=1,random_state=42)

plt.figure(figsize=(16,5))

sns.distplot(train2[train2['target']==0]['var_'+str(0)], label = 't=0')
sns.distplot(train2[train2['target']==1]['var_'+str(0)], label = 't=1')
plt.legend()
plt.yticks([])
plt.xlabel('Var_'+str(0))

plt.show()
```
